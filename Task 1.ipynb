```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("customers-100.csv")
data.head(5)
```

```python
df = pd.DataFrame(data)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop(columns=["Phone 2"],inplace=True)
df.head(5)
```

```python
df.duplicated().sum() 
```

```python
df.duplicated()
```

```python
df.drop_duplicates()
df.head(5)
```

```python
df["Full_Name"] = df["First Name"]+" "+df["Last Name"]
df = df.drop(columns=["First Name","Last Name"])
df
```

```python
col = df.pop("Full_Name")
df.insert(2, "Full_Name", col)
df.head(5)
```

```python
df = df.rename(columns={"Full_Name":"Name"})
df.head(5)
```

```python
df.isna().sum()
```

```python
df.dropna()
```

```python
df["Customer Id"].unique()
```

```python
df['Customer Id'] = df['Customer Id'].astype(str)
df.head(5)
```

```python
df["Name"] = df["Name"].str.lower()
df.head(5)
```

```python
df = df.reindex(sorted(df.columns), axis=1)
df.head(5)
```

```python
cl = df.pop("Index")
df.insert(0,"Index",cl)
df.head(3)
```

```python
df
```

```python
df.to_csv("cleaned_data.csv", index=False)
```
